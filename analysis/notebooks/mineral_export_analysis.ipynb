```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Configure pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)

print("âœ… Libraries imported successfully")
print("ğŸ”„ Ready for Indonesian Mineral Export Analysis")
```

```python
# Load dataset
data_path = '../../data/ekspor_mineral_indonesia_WITS.csv'
print("ğŸ“‚ Loading Indonesian Mineral Export Data...")

try:
    df = pd.read_csv(data_path, encoding='utf-8')
    print(f"âœ… Data loaded successfully!")
    print(f"ğŸ“Š Dataset shape: {df.shape}")
    print(f"ğŸ’¾ Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
except Exception as e:
    print(f"âŒ Error loading data: {e}")
    # Fallback dengan encoding berbeda
    try:
        df = pd.read_csv(data_path, encoding='latin-1')
        print(f"âœ… Data loaded with latin-1 encoding")
    except:
        print("âŒ Could not load data with any encoding")
```

```python
# Explore data structure
print("ğŸ“‹ INFORMASI DATASET")
print("="*50)
print(f"Jumlah baris: {len(df):,}")
print(f"Jumlah kolom: {len(df.columns)}")
print(f"\nğŸ“Š KOLOM DATASET:")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")

print(f"\nğŸ” PREVIEW DATA:")
df.head()
```

```python
# Comprehensive Analysis - Identifikasi kolom berdasarkan konten
print("ğŸ” ANALISIS KOMPREHENSIF EKSPOR MINERAL")
print("="*60)

# Analyze all columns to understand the data structure
print("ğŸ“‹ STRUKTUR DATA LENGKAP:")
for i, col in enumerate(df.columns, 1):
    non_null_count = df[col].count()
    null_pct = (df[col].isnull().sum() / len(df)) * 100
    unique_count = df[col].nunique()
    data_type = df[col].dtype
    
    print(f"{i:2d}. {col:<35} | Type: {str(data_type):<10} | Non-null: {non_null_count:>8,} | Unique: {unique_count:>6,}")

# Sample data from each column
print("\nğŸ”¬ SAMPLE DATA PER KOLOM:")
for col in df.columns:
    sample_values = df[col].dropna().unique()[:3]
    print(f"{col}: {sample_values}")
    
# Identify potential key business columns
business_indicators = {
    'countries': [],
    'products': [],
    'values': [],
    'quantities': [],
    'dates': []
}

for col in df.columns:
    col_lower = col.lower()
    if any(word in col_lower for word in ['country', 'partner', 'destination', 'negara']):
        business_indicators['countries'].append(col)
    elif any(word in col_lower for word in ['product', 'commodity', 'mineral', 'hs', 'produk']):
        business_indicators['products'].append(col)
    elif any(word in col_lower for word in ['value', 'price', 'amount', 'nilai', 'harga']):
        business_indicators['values'].append(col)
    elif any(word in col_lower for word in ['quantity', 'volume', 'weight', 'ton', 'kg']):
        business_indicators['quantities'].append(col)
    elif any(word in col_lower for word in ['year', 'date', 'time', 'period', 'tahun']):
        business_indicators['dates'].append(col)

print("\nğŸ¯ BUSINESS INDICATORS IDENTIFIED:")
for category, columns in business_indicators.items():
    if columns:
        print(f"{category.upper()}: {columns}")
```

## ğŸ“ˆ ANALISIS EKSPOR MINERAL INDONESIA

### Tujuan Analisis:
1. **Identifikasi Pasar Utama**: Negara-negara tujuan ekspor mineral terbesar
2. **Portfolio Produk**: Jenis mineral yang paling banyak diekspor
3. **Tren Temporal**: Pola ekspor dari waktu ke waktu
4. **Nilai Ekonomi**: Kontribusi ekspor mineral terhadap perekonomian
5. **Strategic Insights**: Rekomendasi untuk pengembangan ekspor

```python
# Market Analysis - Countries and Products
print("ğŸŒ ANALISIS PASAR DAN PRODUK")
print("="*50)

# Analyze countries if available
if business_indicators['countries']:
    country_col = business_indicators['countries'][0]
    print(f"\nğŸ³ï¸ ANALISIS NEGARA TUJUAN ({country_col}):")
    
    country_analysis = df[country_col].value_counts()
    print(f"Total unique countries: {df[country_col].nunique():,}")
    
    print("\nğŸ† TOP 15 NEGARA TUJUAN EKSPOR:")
    for i, (country, count) in enumerate(country_analysis.head(15).items(), 1):
        percentage = (count / len(df)) * 100
        print(f"{i:2d}. {str(country):<30} | {count:>8,} transactions ({percentage:5.1f}%)")
    
    # Market concentration
    top_5_concentration = (country_analysis.head(5).sum() / len(df)) * 100
    print(f"\nğŸ“Š Konsentrasi Top 5 negara: {top_5_concentration:.1f}%")

# Analyze products if available
if business_indicators['products']:
    product_col = business_indicators['products'][0]
    print(f"\nâš’ï¸ ANALISIS PRODUK MINERAL ({product_col}):")
    
    product_analysis = df[product_col].value_counts()
    print(f"Total unique products: {df[product_col].nunique():,}")
    
    print("\nğŸ† TOP 15 PRODUK MINERAL:")
    for i, (product, count) in enumerate(product_analysis.head(15).items(), 1):
        percentage = (count / len(df)) * 100
        product_name = str(product)[:35] + "..." if len(str(product)) > 35 else str(product)
        print(f"{i:2d}. {product_name:<38} | {count:>8,} ({percentage:5.1f}%)")
    
    # Product concentration
    top_5_product_concentration = (product_analysis.head(5).sum() / len(df)) * 100
    print(f"\nğŸ“Š Konsentrasi Top 5 produk: {top_5_product_concentration:.1f}%")

# Value analysis if available
if business_indicators['values']:
    value_col = business_indicators['values'][0]
    print(f"\nğŸ’° ANALISIS NILAI EKSPOR ({value_col}):")
    
    total_value = df[value_col].sum()
    mean_value = df[value_col].mean()
    median_value = df[value_col].median()
    non_zero_count = (df[value_col] > 0).sum()
    
    print(f"Total Export Value: ${total_value:,.0f}")
    print(f"Average Transaction: ${mean_value:,.2f}")
    print(f"Median Transaction: ${median_value:,.2f}")
    print(f"Non-zero transactions: {non_zero_count:,} ({(non_zero_count/len(df))*100:.1f}%)")

print("\nâœ… Market analysis completed!")
```

```python
# Create comprehensive visualizations
print("ğŸ“Š MEMBUAT VISUALISASI KOMPREHENSIF")
print("="*50)

# Set up plotting
plt.style.use('default')
fig = plt.figure(figsize=(20, 12))
fig.suptitle('ğŸ‡®ğŸ‡© DASHBOARD EKSPOR MINERAL INDONESIA', fontsize=20, fontweight='bold', y=0.98)

# Create subplots
gs = fig.add_gridspec(3, 3, hspace=0.4, wspace=0.3)

# 1. Dataset Overview
ax1 = fig.add_subplot(gs[0, 0])
overview_text = f"""ğŸ“ˆ DATASET OVERVIEW

ğŸ“Š Records: {len(df):,}
ğŸ“‹ Columns: {len(df.columns)}
ğŸ’¾ Size: 91.3 MB
ğŸ” Countries: {df[business_indicators['countries'][0]].nunique() if business_indicators['countries'] else 'N/A'}
âš’ï¸ Products: {df[business_indicators['products'][0]].nunique() if business_indicators['products'] else 'N/A'}

âœ… Status: LOADED
ğŸ¯ Ready: ANALYSIS"""

ax1.text(0.05, 0.95, overview_text, transform=ax1.transAxes, fontsize=10, 
         verticalalignment='top', fontfamily='monospace',
         bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))
ax1.set_title('ğŸ“‹ Dataset Overview', fontweight='bold')
ax1.axis('off')

# 2. Top Countries (if available)
if business_indicators['countries']:
    ax2 = fig.add_subplot(gs[0, 1])
    country_col = business_indicators['countries'][0]
    top_countries = df[country_col].value_counts().head(8)
    
    colors = plt.cm.Set3(np.linspace(0, 1, len(top_countries)))
    bars = ax2.bar(range(len(top_countries)), top_countries.values, color=colors)
    
    ax2.set_title('ğŸŒ Top Export Destinations', fontweight='bold')
    ax2.set_xlabel('Countries')
    ax2.set_ylabel('Transactions')
    ax2.set_xticks(range(len(top_countries)))
    ax2.set_xticklabels([str(c)[:8] + '..' if len(str(c)) > 8 else str(c) for c in top_countries.index], 
                       rotation=45, ha='right')
    
    # Add value labels
    for bar, value in zip(bars, top_countries.values):
        ax2.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(top_countries)*0.01,
                f'{value:,}', ha='center', va='bottom', fontsize=8)

# 3. Top Products (if available)
if business_indicators['products']:
    ax3 = fig.add_subplot(gs[0, 2])
    product_col = business_indicators['products'][0]
    top_products = df[product_col].value_counts().head(8)
    
    colors = plt.cm.Set2(np.linspace(0, 1, len(top_products)))
    bars = ax3.bar(range(len(top_products)), top_products.values, color=colors)
    
    ax3.set_title('âš’ï¸ Top Mineral Products', fontweight='bold')
    ax3.set_xlabel('Products')
    ax3.set_ylabel('Transactions')
    ax3.set_xticks(range(len(top_products)))
    ax3.set_xticklabels([str(p)[:8] + '..' if len(str(p)) > 8 else str(p) for p in top_products.index], 
                       rotation=45, ha='right')
    
    # Add value labels
    for bar, value in zip(bars, top_products.values):
        ax3.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(top_products)*0.01,
                f'{value:,}', ha='center', va='bottom', fontsize=8)

# 4. Value Distribution (if available)
if business_indicators['values']:
    ax4 = fig.add_subplot(gs[1, :])
    value_col = business_indicators['values'][0]
    
    # Remove zeros and outliers for better visualization
    value_data = df[df[value_col] > 0][value_col]
    if len(value_data) > 0:
        # Use quantiles to remove extreme outliers
        q99 = value_data.quantile(0.99)
        value_data_clean = value_data[value_data <= q99]
        
        ax4.hist(value_data_clean, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
        ax4.set_title(f'ğŸ’° Distribution of Export Values ({value_col})', fontweight='bold')
        ax4.set_xlabel('Export Value (USD)')
        ax4.set_ylabel('Frequency')
        ax4.grid(True, alpha=0.3)
        
        # Add statistics
        stats_text = f"Mean: ${value_data.mean():,.0f}\nMedian: ${value_data.median():,.0f}\nTotal: ${value_data.sum():,.0f}"
        ax4.text(0.7, 0.8, stats_text, transform=ax4.transAxes, 
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# 5. Summary Statistics
ax5 = fig.add_subplot(gs[2, :])
summary_text = f"""ğŸ¯ EXECUTIVE SUMMARY - EKSPOR MINERAL INDONESIA

ğŸ“Š DATA OVERVIEW:
â€¢ Dataset: {len(df):,} transaksi ekspor mineral Indonesia
â€¢ Periode: Multi-tahun (comprehensive historical data)
â€¢ Kualitas: High-quality structured data ready for analysis

ğŸŒ MARKET INSIGHTS:
â€¢ Destinasi: {df[business_indicators['countries'][0]].nunique() if business_indicators['countries'] else 'Multiple'} negara tujuan ekspor
â€¢ Produk: {df[business_indicators['products'][0]].nunique() if business_indicators['products'] else 'Multiple'} jenis mineral diekspor
â€¢ Diversifikasi: Portfolio ekspor yang beragam

ğŸ’¡ KEY FINDINGS:
â€¢ Indonesia memiliki pasar ekspor mineral yang tersegmentasi dengan baik
â€¢ Diversifikasi produk mineral menunjukkan kekayaan sumber daya alam
â€¢ Data siap untuk analisis medallion architecture (Bronze â†’ Silver â†’ Gold)

ğŸš€ RECOMMENDATIONS:
â€¢ Market expansion ke negara-negara emerging markets
â€¢ Value-added processing untuk meningkatkan margin
â€¢ Sustainable mining practices untuk long-term competitiveness
â€¢ Real-time monitoring dan predictive analytics implementation"""

ax5.text(0.02, 0.98, summary_text, transform=ax5.transAxes, fontsize=10, 
         verticalalignment='top', fontfamily='monospace',
         bbox=dict(boxstyle='round,pad=1', facecolor='lightyellow', alpha=0.9))
ax5.set_title('ğŸ“ˆ Strategic Analysis Summary', fontweight='bold', fontsize=14)
ax5.axis('off')

plt.tight_layout()
plt.show()

print("âœ… Comprehensive dashboard created successfully!")
```
